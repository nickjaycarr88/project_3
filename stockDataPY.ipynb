```python
# Storing datetime value for 1 year ago from today
# Time Period in Years
## Change the value here to increase or decrease the data peroid wanted
timePeriod = 1

numDays = timePeriod * 365

startDate = dt.datetime.now() - dt.timedelta(days=numDays)
startDate = startDate.date()
print(startDate)
```

```python
nowDate = dt.datetime.now().date()
print(nowDate)
```
